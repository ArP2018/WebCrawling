```python
import re
from pandas import DataFrame
def find_fakuan(text):
     if '罚款' in str(text):
#         print('index: ' + str(i))
        ret = re.findall(re.compile('(罚款[^罚^元]*元)'), text)
        if ret:
            print(ret[0])
            return ret[0]
        else:            
            ret = re.findall(re.compile('(处[^处]*罚款)'), text)
            if ret:
                print(ret[0])
                return ret[0]
            else:
                print(text)
                return text
        return ''
```
