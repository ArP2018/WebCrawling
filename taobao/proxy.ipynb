```python
import requests
from bs4 import BeautifulSoup
url = 'https://www.xicidaili.com/nn/'
header = {
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/68.0.3423.2 Safari/537.36'
}
resp = requests.get(url, headers=header)
soup = BeautifulSoup(resp.content, 'lxml')

```

```python
from requests.exceptions import ProxyError
import traceback
def check_proxy(ip, port):
    url = 'http://www.baidu.com'
    try:
        resp = requests.get(url, headers=header, proxies={'http':'http://{0}:{1}'.format(ip, port)}, timeout=10, verify=False)
        if resp.status_code == 200:
            print('available proxy: {0}:{1}'.format(ip, port))
    except TimeoutError:
        print('connection timed out')
    except ProxyError:
        print('proxy error')
    except:
        print('other error')
```

```python
for r in soup.find('table', id='ip_list').find_all('tr')[1:]:
    tds = r.find_all('td')
    ip, port, proxy_type = tds[1].text, tds[2].text, tds[5].text
    if proxy_type.upper() == 'HTTP':
        print(ip, port)
        
        check_proxy(ip, port)
```

```python
check_proxy('', '22752')
```

```python
# 快代理
url = 'https://www.kuaidaili.com/ops/'

k_resp = requests.get(url, headers=header)
k_soup = BeautifulSoup(k_resp.text, 'lxml')
```

```python
for tr in k_soup.find('div', id='freelist').find_all('tr')[1:]:
    tds = tr.find_all('td')
    ip,port, proxy_type = tds[0].text, tds[1].text, tds[3].text
    print(ip, port, proxy_type)
    if 'HTTPS' in proxy_type.upper():
        check_proxy(ip, port)
```
