```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'http://list.tmall.com/search_product.htm?q=%CF%B4%B7%A2%CB%AE&type=p&vmarket=&spm=a2233.7711963.a2227oh.d100&from=..pc_1_searchbutton'
```

```python
resp = requests.get(url, headers=header)
```

```python
soup = BeautifulSoup(resp.content, 'lxml')
```

```python
soup.body
```

```python
soup.find_all('div', class_='product ')[2]
```

```python
product_list = soup.find_all('div', class_='product ')
for p in product_list:
    product_id = p.attrs['data-id']
    title = p.find('p', class_='productTitle').find('a').attrs['title']
    url = p.find('p', class_='productTitle').find('a').attrs['href'].lstrip('//')
    shop = p.find('div', class_='productShop').find('a').text
    print(product_id, title, url, shop)
```
