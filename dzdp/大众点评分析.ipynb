```python
# by 门店
review_all_url = 'http://www.dianping.com/shop/32500835/review_all/p'

for i in range(1, 101):
    resp = requests.get(review_all_url+str(i), headers=header)
    soup = BeautifulSoup(resp.text, 'lxml')
    reviews = soup.find_all('div', class_='main-review')
    for r in reviews:
        user_name = r.find('div', class_='dper-info').find('a').text.strip()
        score = r.find('span', class_='score').text.strip()
        desc_hide = r.find('div', class_='review-words Hide')
        if desc_hide:
            desc = desc_hide.text
        else:
            desc = r.find('div', class_='review-words').text.strip()
        
        print('user_name: '+user_name)
        print('score: '+score)
        print(desc)
        print('-'*100)
        
        time.sleep(5)
            
    
```
